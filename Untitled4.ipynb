<strong>Dark Channel Prior method: </strong>\
The dark channel of an image is defined as the minimum intensity value within a local window. The haze in an image tends to increase the intensity of pixels, making the dark channel value of the image smaller. Therefore, the dark channel prior method is used to estimate the atmospheric light and transmission map of the image. The atmospheric light is estimated by selecting the top 0.0001% brightest pixels in the image, while the transmission map is obtained by applying an erosion operation to the dark channel of the image.

```python
# Define function for finding the dark channel of an image
def dark_channel(img, size=15):
    # Split the input image into its RGB channels
    r, g, b = cv2.split(img)
    # Take the minimum of the RGB channels to get a grayscale image
    min_img = cv2.min(r, cv2.min(g, b))
    # Create a rectangular kernel for erosion
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (size, size))
    # Erode the minimum image using the kernel to find the dark channel
    dc_img = cv2.erode(min_img, kernel)
    # Return the dark channel image
    return dc_img
```

<strong>Atmospheric Light Estimation</strong>\
The atmospheric light is the maximum intensity value in the image, and it is assumed to be the same for all pixels in the scene. This assumption is based on the observation that light scattering due to haze affects all colors equally, so the brightest part of the image is likely to be the least affected by haze. The atmospheric light is estimated by taking the mean intensity value of the top percent of pixels in the dark channel.

```python
# Define function for finding the atmospheric light of an image
def get_atmo(img, percent=0.0001):
    # Compute the mean pixel value for each pixel in the image
    mean_perpix = np.mean(img, axis=2).reshape(-1)
    # Take the top percent of pixel means to estimate the atmospheric light
    mean_topper = mean_perpix[:int(img.shape[0] * img.shape[1] * percent)]
    # Return the mean of the top pixel means as the atmospheric light
    return np.mean(mean_topper)
```

<strong>Transmission Map Estimation</strong>\
The transmission map is estimated using the atmospheric light and the dark channel prior. Specifically, the transmission map is defined as 1 - w times the dark channel of the image, where w is a constant that controls the strength of the haze removal. The guided filter is applied to the transmission map to smooth out noise and preserve edges.

```python
# Define function for finding the transmission of an image
def get_trans(img, atom, w=0.3):
    # Divide the image by the atmospheric light to get the initial estimate of transmission
    x = img / atom
    # Use the dark channel of the transmission estimate to refine the transmission
    t = 1 - w * dark_channel(x, 5)
    # Return the final transmission estimate
    return t
```

<strong>Guided Filter</strong>\
The guided filter is a non-linear image filtering technique that is used to smooth an image while preserving edges. It is a generalization of the bilateral filter, which is a linear filter that smooths an image while preserving edges based on the difference in intensity values between neighboring pixels. The guided filter uses a guidance image to compute the filtering coefficients, which are then used to filter the input image. In this code, the input image is the transmission map and the guidance image is the grayscale version of the input image after contrast enhancement. The guided filter is used to smooth the transmission map and remove noise, while preserving the edges of objects in the scene.

```python
# Define function for applying a guided filter to an image
def guided_filter(p, i, r, e):
    """
    :param p: input image
    :param i: guidance image
    :param r: radius
    :param e: regularization
    :return: filtering output q
    """
    # Compute the mean and correlation matrices for the guidance image and input image
    mean_I = cv2.boxFilter(i, cv2.CV_64F, (r, r))
    mean_p = cv2.boxFilter(p, cv2.CV_64F, (r, r))
    corr_I = cv2.boxFilter(i * i, cv2.CV_64F, (r, r))
    corr_Ip = cv2.boxFilter(i * p, cv2.CV_64F, (r, r))
    # Compute the variances and covariance from the mean and correlation matrices
    var_I = corr_I - mean_I * mean_I
    cov_Ip = corr_Ip - mean_I * mean_p
    # Compute the filter parameters a and b
    a = cov_Ip / (var_I + e)
    b = mean_p - a * mean_I
    # Compute the mean of a and b over a local region using a rectangular kernel
    mean_a = cv2.boxFilter(a, cv2.CV_64F, (r, r))
    mean_b = cv2.boxFilter(b, cv2.CV_64F, (r, r))
    # Compute the output image q using the filter parameters and input image
    q = mean_a * i + mean_b
    # Return the filtered output image
    return q
```

```python
def dehaze(img_path, w=0.3, tmin=0.3, r=20, e=0.0001):
    im = cv2.imread(img_path)
    
    #converts the image to a floating-point format and scales the pixel values to the range [0, 1].
    img = im.astype('float64') / 255 
    
    #converts the image to grayscale and scales the pixel values to the range [0, 1]
    img_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY).astype('float64') / 255
    
    
    #estimates the atmospheric light in the image using the get_atmo function, which computes the brightest pixels in the image
    atom = get_atmo(img)
    
    #estimates the transmission map using the get_trans function, 
    #which computes the minimum value of each local patch in the dark channel of the image.
    trans = get_trans(img, atom, w=w)

    # Avoid over-darkening the image by setting a minimum value for transmission
    trans = np.maximum(trans, tmin)
    
    
    #refines the transmission map using the guided filter, which smooths the transmission map while preserving edges.
    trans_guided = guided_filter(trans, img_gray, r=r, e=e)
    
    #nitializes an array for the dehazed image with the same size as the original image.
    result = np.empty_like(img)
    
    # loops over the three color channels (red, green, and blue)
    for i in range(3):
        result[:, :, i] = (img[:, :, i] - atom) / trans_guided + atom
        #computes the dehazed image for each color channel by dividing the difference between the original image and 
        #the atmospheric light by the refined transmission map, and then adding the atmospheric light back in

    return result
```

<strong>Contrast Limited Adaptive Histogram Equalization (CLAHE)</strong>\
The "enhance_back" function is a post-processing step used to enhance the contrast of a dehazed image. To do this, it uses a technique called Contrast Limited Adaptive Histogram Equalization (CLAHE).

CLAHE is a technique that redistributes the intensity values of an image, which can improve the visibility of details and features that may be obscured by haze. In the case of the "enhance_back" function, CLAHE is applied to the L channel of the LAB color space.

The LAB color space is a color model that separates the luminance (L) from the chrominance (A and B) of an image. The L channel represents the brightness of an image, while the A and B channels represent the color information.

In this function, the L channel is extracted from the dehazed image, and histogram equalization is applied to it. Histogram equalization is a method of image processing that redistributes the intensity values of an image so that they span the full range of possible values. This can be useful in enhancing the contrast of an image.

By applying CLAHE to the L channel of the LAB color space, the "enhance_back" function can improve the contrast of the dehazed image, making it more visually appealing and easier to analyze. This can be especially useful in situations where important details or features are obscured by haze, as it can help make them more visible.

```python
def enhance_back(img):
    # Convert image to LAB color space
    img = np.clip(img * 255, 0, 255).astype(np.uint8)  # Scale and convert to 8-bit
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

    # Split channels
    l, a, b = cv2.split(lab)

    # Apply histogram equalization to the L channel
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    l = clahe.apply(l)

    # Merge channels back together
    lab = cv2.merge((l, a, b))

    # Convert back to RGB color space
    enhanced = cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)

    return enhanced
```
